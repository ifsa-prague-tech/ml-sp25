```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_covtype
from sklearn.metrics import RocCurveDisplay, roc_auc_score, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
RANDOM_SEED = 131415

TEST_SIZE=0.80
```

#### Loading dataset

Forest Cover Types dataset

```python
cover = fetch_covtype()
```

```python
print(cover.DESCR)
```

```python
cover.target_names
```

```python
cover.target = cover.target-1
```

```python
set(cover.target)
```

```python
class_names = {
    0: "0: Lodgepole Pine",
    1: "1: Spruce/Fir",
    2: "2: Ponderosa Pine",
    3: "3: Cottonwood/Willow",
    4: "4: Aspen",
    5: "5: Douglas-fir",
    6: "6: Krummholz"
}
```

```python
cover.feature_names
```

```python
df = pd.DataFrame(cover.data, columns=cover.feature_names)
df['Cover_Type'] = cover.target
df['Cover_Type_Desc'] = df['Cover_Type'].map(class_names)
```

```python
# Create a histogram for 'Elevation' by 'Cover_Type'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Elevation", hue="Cover_Type_Desc", element="step", palette="tab10", multiple="stack", bins=30)

# Add labels and title
plt.title("Elevation Histogram by Cover Type")
plt.xlabel("Elevation")
plt.ylabel("Count")

plt.show()
```

```python
# Create a density plot for 'Elevation' by 'Cover_Type'
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="Elevation", hue="Cover_Type_Desc", fill=True, palette="tab10")

# Add labels and title
plt.title("Elevation Density Plot by Cover Type")
plt.xlabel("Elevation")
plt.ylabel("Density")

plt.show()
```

```python
X_train, X_dev, y_train, y_dev = train_test_split(
    cover.data, cover.target, random_state=RANDOM_SEED, test_size=TEST_SIZE)
```
